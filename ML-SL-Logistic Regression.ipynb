#SALE PREDICTION FROM EXISTING CUSTOMNER using LOGISTIC REGRSSION


```python
import pandas as pd 
import numpy as np
```

```python
data=pd.read_csv("Digitalad_dataset.csv")
```

```python
print(data.shape)
print(data.head(5))
```

```python
x=data.iloc[:,:-1].values
x
```

```python
y=data.iloc[:,-1:].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(random_state=0)
model.fit(x_train,y_train)
```

```python
age=int(input("Ente the customer's age:"))
salary=int(input("Enter the customer's salary:"))
newcustomer=[[age,salary]]
result=model.predict(sc.transform(newcustomer))
print(result)
if result==1:
    print("The customer will buy")
else:
    print("The customer won't buy")
```

```python
y_pred=model.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
print("Accuracy of the model is {0}%".format(accuracy_score(y_test,y_pred)*100))
```
