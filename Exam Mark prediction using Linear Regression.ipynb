```python
#import libaries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#load dataset
dataset=pd.read_csv('mark.csv')
```

```python
#summarize dataset
print(dataset.shape)
print(dataset.head(5))
```

```python
#finding and removing NAN values
dataset.columns[dataset.isna().any()]
```

```python
dataset.hours=dataset.hours.fillna(dataset.hours.mean())
```

```python
#Segregate dataset
X=dataset.iloc[:,:-1].values
X
```

```python
Y=dataset.iloc[:,-1].values
Y
```

```python
#Training dataset using Linear Regression
model=LinearRegression()
model.fit(X,Y)
```

```python
#Predicted mark for custom values
hrs=float(input("Enter the number of hours:"))
student_age=int(input("Enter the age of student:"))
net=int(input("Enter the number 0-no & 1-yes:"))
mark=[[hrs,student_age,net]]
markresult=model.predict(mark)
print("Predicted mark of the student is:",markresult)
```
