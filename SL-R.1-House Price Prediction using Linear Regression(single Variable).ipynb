```python
#import libaries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#load dataset
df=pd.read_csv('dataset.csv')
```

```python
#summarize
print(df.shape)
print(df.head(5))
```

```python
#visualize dataset
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='*')
```

```python
#segregate dataset
X=df.drop('price',axis='columns')
X
```

```python
Y=df.price
Y
```

```python
#Training dataset
model=LinearRegression()
model.fit(X,Y)
```

```python
#predicted price for land sq.feet of custom values
x=int(input("Enter the area in sq.ft:"))
landareasqft=[[x]]
predictedmodelresult=model.predict(landareasqft)
print("predicted price of {0}sq.ft is {1}".format(x,predictedmodelresult))
```

```python
#Let's ckeck our model is right?
#y=mx+c
m=model.coef_
print(m)
```

```python
b=model.intercept_
print(b)
```

```python
y=m*x+b
print("The price of {0}sq.ft area is:{1}".format(x,y[0]))
```
