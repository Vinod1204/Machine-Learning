```python
#importing libaries 
import pandas as pd
```

```python
#load dataset
df=pd.read_csv('cardataset.csv')
df
```

```python
#dropping the car_id
df=df.drop(['car_ID'],axis=1)
df
```

```python
#summarize dataset
print(df.shape)
print(df.head(5))
```

```python
#Splitting dataset
Xdata=df.drop('price',axis='columns')
numericalcols=Xdata.select_dtypes(exclude=['object']).columns
X=Xdata[numericalcols]
X
```

```python
Y=df['price']
Y
```

```python
#Scaling the independent variable
from sklearn.preprocessing import scale
cols = X.columns
X = pd.DataFrame(scale(X))
X.columns = cols
X
```

```python
#Splitting dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.27,random_state=0)
```

```python
#Training using Random Forest
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100,criterion='mse',max_depth=35,max_features="sqrt",min_samples_split=2,n_jobs=-1,
                             random_state=0,max_samples=None)
model.fit(x_train,y_train)
```

```python
#Evaluating Model
ypred=model.predict(x_test)
from sklearn.metrics import r2_score
r2score=r2_score(y_test,ypred)
print("R2score:",r2score*100)
```
